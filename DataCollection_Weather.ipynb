```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
```

```python
races = pd.read_csv('races.csv')
```

```python
#create table for weather 
weather = races.iloc[:, [0,1,2,3,7]]
weather['weather_det'] = [None] * len(weather)
```

```python
#scrape each wiki url for weather information 
for index, row in weather.iterrows(): 
    url = row['url']
    r = requests.get(url, verify=False)
    #find weather table 
    soup = BeautifulSoup(r.text, 'html.parser')
    table = soup.find('table', {'class': 'infobox'})
    df = pd.read_html(str(table))
    df = pd.DataFrame(df[0])
    df = df.set_index(df.columns[0])
    #set weather wiki value for weather column
    try: 
        weather.at[index,'weather_det'] = df.loc['Weather', :][1]
    except: 
        weather.at[index,'weather_det'] = "None"
```

```python
# set up a dictionary to convert weather details into keywords
weather_dict = {'weather_warm': ['soleggiato', 'clear', 'warm', 'hot', 'sunny', 'fine', 'mild', 'sereno'],
               'weather_cold': ['cold', 'fresh', 'chilly', 'cool'],
               'weather_dry': ['dry', 'asciutto'],
               'weather_wet': ['showers', 'wet', 'rain', 'pioggia', 'damp', 'thunderstorms', 'rainy'],
               'weather_cloudy': ['overcast', 'nuvoloso', 'clouds', 'cloudy', 'grey', 'coperto']}

#define function to return weather keyword
def retWeather(x):
    weather_arr = []
    for k, v in weather_dict.items(): 
        if any(el in set(v) for el in x):
            weather_arr.append(k)
    return weather_arr
```

```python
#apply retWeather function to create weather column with keywords
weather['weatherDetail'] = weather['weather_det'].apply(lambda x: x.lower().split() if x else 'None')
weather['weather'] = weather['weatherDetail'].apply(retWeather)
```

```python
#export weather data as csv file
weather.to_csv('weather.csv')
```
